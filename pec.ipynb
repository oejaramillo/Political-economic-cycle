```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir un rango de valores para tau (impuestos) y g (bienes pblicos)
tau_values = np.linspace(0, 10, 100)
y = 10  # Ingreso total disponible
epsilon = 2  # Trmino adicional en la restriccin

# Funciones de utilidad U(y - tau) y V(tau + epsilon - g), asumiendo formas funcionales simples
U = lambda tau: np.sqrt(y - tau)  # Raz cuadrada para U, que es cncava
V = lambda g: np.sqrt(g + epsilon)  # Raz cuadrada para V, que es cncava

# Calcular g a partir de la restriccin tau + epsilon - g
g_values = tau_values + epsilon

# Curvas de indiferencia para U y V
U_values = U(tau_values)
V_values = V(g_values)

# Generar valores aleatorios para alpha_t^i y alpha_t-1^i
np.random.seed(0)  # Para reproducibilidad
num_samples = 100
alpha_t = np.random.binomial(1, 0.5, num_samples)
alpha_t_minus_1 = np.random.binomial(1, 0.3, num_samples)

# Calcular epsilon_t^i
epsilon_t_i = alpha_t + alpha_t_minus_1

# Recalcular V para diferentes valores de epsilon_t^i
V_values_epsilon = [np.sqrt(g_values + epsilon + epsilon_t) for epsilon_t in epsilon_t_i]

# Graficar con diferentes realizaciones de epsilon_t^i
plt.figure(figsize=(10, 6))
plt.plot(tau_values, U_values, label='Utilidad privada U(y - )', color='blue')

# Graficar varias curvas de V para diferentes epsilon_t^i
for i, V_val in enumerate(V_values_epsilon[:10]):  # Solo graficamos algunas para evitar saturacin
    plt.plot(g_values, V_val, label=f'Utilidad pblica V(+-g) = {epsilon_t_i[i]}' if i == 0 else "", color='orange', alpha=0.5)

plt.xlabel('Consumo Privado (y - ) o Bienes Pblicos (g)')
plt.ylabel('Utilidad')
plt.title('Grfico con Variable Aleatoria _t^i')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Generar valores aleatorios para alpha_t^i y alpha_t-1^i para nuevas realizaciones de epsilon_t^i
np.random.seed(1)  # Cambio de semilla para nuevas realizaciones
alpha_t_new = np.random.binomial(1, 0.5, num_samples)
alpha_t_minus_1_new = np.random.binomial(1, 0.5, num_samples)

# Calcular nuevas realizaciones de epsilon_t^i
epsilon_t_i_new = alpha_t_new + alpha_t_minus_1_new

# Definir rangos para g (bienes pblicos) y tau (impuestos)
g_values = np.linspace(0, 10, 100)
tau_values = np.linspace(0, 10, 100)

# Funciones de utilidad simplificadas
U = lambda tau: np.sqrt(y - tau)
V = lambda tau, g, epsilon: np.sqrt(tau + epsilon - g)

# Graficar para diferentes realizaciones de epsilon_t^i
plt.figure(figsize=(10, 6))

for epsilon_t in epsilon_t_i_new[:3]:  # Solo graficamos algunas para evitar saturacin
    W_values = U(tau_values) + V(tau_values, g_values, epsilon_t)
    plt.plot(g_values, W_values, label=f'W con  = {epsilon_t}', alpha=0.5)

plt.xlabel('Inversin pblica (g)')
plt.ylabel('Impuestos ()')
plt.title('Funciones de utilidad W para  binomial (p=0.5)')
plt.text(2, 1.5, 'Lder ms competente,\n(mayor ) utilidad crece hacia arriba', fontsize=9, color='black')
plt.legend()
plt.grid(True)
plt.show()
```
